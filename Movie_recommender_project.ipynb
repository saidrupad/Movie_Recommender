Next step is to bring those columns in to correct format

```python
movies.iloc[0].genres
```

Our goal is to bring [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] 

to   ['Action', 'Adventure', 'Fantasy', 'Science Fiction']

```python
def convert(obj):
    List = []
    for i in obj:
        List.append(i['name'])
    return List
```

```python
convert(movies['genres'])
```

Unable to pass in to function because right now movies['genres'] is a string and the function is expecting a list
so to change string to list there is a library called ast

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# In cast , we want only top 3 cast names
def convert3(obj):
    List = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            List.append(i['name'])
            cnt+=1
        else :
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' :
            List.append(i['name'])
    return List
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Converting overview string to List using lamda function
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies["genres"] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies["keywords"] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies["cast"] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies["crew"] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title' , 'tags']]
```

Converting list back to string

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

Stemming : [loved , loving, love] => [love , love , love]

           [activities , activity] => [activity , activity]

```python
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

To not loose track of indices

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse = True , key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(f"{new_df.iloc[i[0]].title} {i[0]}")
        
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
